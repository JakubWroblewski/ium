# Dane

```python
df = pd.read_csv("../data/final_data.csv", delimiter=';')
df.drop(columns=['id'], inplace=True)
df.head()
```

```python
train=df.sample(frac=0.8,random_state=205)
test=df.drop(train.index)

X_train = torch.tensor(train.values[:, :-1], dtype=torch.float32)
y_train = torch.tensor(train.values[:, -1], dtype=torch.float32)

X_test = torch.tensor(test.values[:, :-1], dtype=torch.float32)
y_test = torch.tensor(test.values[:, -1], dtype=torch.float32)

train_dataset = data.TensorDataset(X_train, y_train)
test_dataset = data.TensorDataset(X_test, y_test)

train_loader = data.DataLoader(train_dataset, batch_size=64, shuffle=True, pin_memory=True)
test_loader = data.DataLoader(test_dataset, batch_size=64, shuffle=False, pin_memory=True)
```
